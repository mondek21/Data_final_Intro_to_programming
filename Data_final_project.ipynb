```python
#to make column titles more accurate.....
#change column name for "month" to "date"
gun_data.rename(columns={'month': 'date'}, inplace=True)
```

```python
#look at first few lines of data to ensure change took place
gun_data.head()
```

```python
#group data by date and include totals with date   # variable swap grouped_data_1 for 
gun_data_grouped_by_date = gun_data.groupby('date').mean()['totals']
gun_data_grouped_by_date.head()
```

```python
#use describe function to get general understanding of new variable
print gun_data_grouped_by_date.describe()
```

```python
#time series of gun system requests over time
%pylab inline
gun_data_grouped_by_date.plot()
```

```python
#this times series looks pretty level from 1998 to 2006.  In 2007 the gun sales start to climb consistently
#with a few spikes throughout that climb
#graph looked cyclical, perhaps on an annual basis, so I want to identify if gun sales vary by month
#to do so, we need to create a new column:  "month"
#add column for monthly data
gun_data['month']= gun_data['date'].str[-2:]
gun_data.head()
```

```python
#group by month and examine how the totals change by month   #swapped grouped_data_2 for 
gun_data_grouped_by_month = gun_data.groupby('month').mean()['totals']
gun_data_grouped_by_month.head()
```

```python
#time series of total gun requests per year
%pylab inline
gun_data_grouped_by_month.plot()
```

```python
#the shape of this graph matches the cyclical behavior of the timeseries graph.  
#it looks like gun sales spike in december, perhaps coinciding with the holidays, perhaps coinciding with hunting
#seasons (in Illinois,there isn't much hunting in the summer.  September has dove season, 
#Nov and Dec host deer season, and March is when turkey season starts
```

```python
#I now want to group data by state, so I can compare results to census data   
#swap grouped_data_3 for gun_data_grouped_by_state 
gun_data_grouped_by_state = gun_data.groupby('state').mean()
gun_data_grouped_by_state.head()
```

```python
#look at basic statistics for data grouped by state
gun_data_grouped_by_state.describe()
```

```python
#read US_Census_data dataset into jupyter notebook with read_csv function
US_Census_Data = pd.read_csv('C:\Users\Mike\Data_Analysis\Course_files\Data_final\US_Census_Data.csv')
```

```python
#examine first few lines of Census data.  I see I will need to change format to perform a merge
US_Census_Data.head()
```

```python
#remove Fact Note column so transposed data will format correctly
del US_Census_Data['Fact Note']
```

```python
#perform transpose function to improve format of Census data  # swapped df_1 for census_data_transposed
census_data_transposed = US_Census_Data.transpose()
```

```python
#examine first few lines of Census Data
census_data_transposed.head()
```

```python
#Upon reviewing format of new dataFrame, I see that formatting is poor.  the first row of data
#is actually the column titles.  I will create a new dataFrame to correct formatting error
#swap df_2 for census_data_transposed_correct_column
headers = census_data_transposed.iloc[0]
census_data_transposed_correct_column = pd.DataFrame(census_data_transposed.values[1:], \
                                                     index = census_data_transposed.index[1:],columns = headers)
```

```python
#display first few lines of df_2
census_data_transposed_correct_column.head()
```

```python
#since the data looks correct now, I will perform a merge. 
#swapped merged_data for census_and_gun_data
census_and_gun_data = pd.merge(census_data_transposed_correct_column, gun_data_grouped_by_state, \
                       left_index=True, right_index=True)
```

```python
#View the first few lines of merged data
census_and_gun_data.head()
```
